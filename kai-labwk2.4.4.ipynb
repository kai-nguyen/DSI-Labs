```python
import pandas as pd
import numpy as np

df_a = pd.DataFrame([{"a": 1, "b": 4}, {"a": 2, "b": 5}, {"a": 3, "b": 6}, 
                    {"c": 1}])
df_a
```

```python
df_b = pd.DataFrame([{"c": 2, "d": 7}, {"c": 3, "d": 8}])
df_b
```

```python
# left join
pd.merge(df_a, df_b, how='left') # okay to omit the 'on', but is good practice to use
```

```python
# right join
pd.merge(df_a, df_b, on='c', how='right')
```

```python
#inner join
pd.merge(df_a, df_b, on='c', how='inner')
```

```python
#outer join
pd.merge(df_a, df_b, on='c', how='outer')
```

```python
# Bonus

left = pd.DataFrame({'key1': ['foo', 'foo', 'bar'],
                  'key2': ['one', 'two', 'one'],
                  'lval': [1, 2, 3]})
left
```

```python

right = pd.DataFrame({'key1': ['foo', 'foo', 'bar', 'bar'],
                   'key2': ['one', 'one', 'one', 'two'],
                   'rval': [4, 5, 6, 7]})
right
```

```python
#left join
pd.merge(left, right, how='left')
pd.merge(left, right, on='key1', how='left')
```

```python
#right join
pd.merge(left, right, on='key1', how='right')
```

```python
#inner join
pd.merge(left, right, on='key1', how='inner')
```

```python
#outer join
pd.merge(left, right, on='key1', how='outer')
```

```python
import pandas as pd
movie_url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.item'
movie_cols = ['movie_id', 'title']
movies = pd.read_table(movie_url, sep='|', header=None, names=movie_cols, usecols=[0, 1])
movies.head()
```

```python
rating_url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.data'
rating_cols = ['user_id', 'movie_id', 'rating', 'timestamp']
ratings = pd.read_table(rating_url, sep='\t', header=None, names=rating_cols)
ratings.head()
```

```python
movie_ratings = pd.merge(movies, ratings, on='movie_id')
movie_ratings.head()
```

```python
movie_ratings[movie_ratings.title=="Star Wars (1977)"].rating.mean()
movie_ratings[movie_ratings.title.str.contains("The")].rating.mean()
```

```python
users_url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user'
users_info = pd.read_table(users_url, sep='|')
users_info.head()
```

```python
users_movies = pd.merge(movie_ratings, users_info, on='user_id', how='left')
users_movies
```

```python
#What's the average (mean) rating for the movie with movie_id 7?
movie_ratings[movie_ratings.movie_id==7].rating.mean()
```

```python
# What's the average (mean) rating for the movie with the title "Star Wars (1977)"

movie_ratings[movie_ratings.title=="Star Wars (1977)"].rating.mean()
```
